```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def customer_based_example1():
    dummy_data = {"Categories": ["Drinks", "Meat/Deli", "Fruit & Veg", "Dairy", "Pet", "Entertainment", "Baby"
                                , "Snacks", "Freezer"], "Number of purchase":[13, 12, 15, 10, 7, 1, 2, 0, 1]}
    df = pd.DataFrame(dummy_data)
    df.plot.bar(x = "Categories", y = "Number of purchase")
```

```python
def customer_based_example2():
    Months =["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
    Total_amount = [350.4, 368.9, 401, 336.7, 349.6, 352, 358, 57.5, 12, 0, 0, 0]
    plt.plot(Months, Total_amount)
    plt.xlabel('Months')
    plt.ylabel("Total amount of purchase ($)")
    plt.title("Customer's amount of purchase by month")
    plt.show()
```
